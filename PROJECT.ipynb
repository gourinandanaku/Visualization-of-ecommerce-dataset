```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris=pd.read_csv("ecommerce_dataset_updated.csv")
plt.figure(figsize=(10,6))
sns.scatterplot(data=iris,x="Price (Rs.)",y="Final_Price(Rs.)",hue="Category",palette="Spectral")
plt.title("Price vs Final_Price")
plt.xlabel("Price(Rs)")
plt.ylabel("Final Price(Rs.)")
plt.legend(title="Category")
plt.show()
```

```python
import pandas as pd
import seaborn as sd
import matplotlib.pyplot as plt
iris=pd.read_csv("ecommerce_dataset_updated.csv")
plt.figure(figsize=(9,4))
sd.lineplot(data=iris,x="Price (Rs.)",y="Final_Price(Rs.)",hue="Category",marker="o")
plt.title("Price(RS) vs Final_Price")
plt.xlabel("Price(Rs)")
plt.ylabel("Final Price(Rs)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris=pd.read_csv("ecommerce_dataset_updated.csv")
avg_sepal_length=iris.groupby("Category")["Price (Rs.)"].mean().reset_index()
plt.figure(figsize=(10,3))
sns.barplot(data=avg_sepal_length,x="Category",y="Price (Rs.)",palette="icefire")
plt.title("Average Price by Category")
plt.xlabel("Category")
plt.ylabel("Price (Rs.)")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))
sns.histplot(data=iris,x="Price (Rs.)",hue="Category",palette="Spectral",bins=15)
plt.title("Distribution of Price in Categories")
plt.xlabel("Price(Rs.)")
plt.ylabel("Frequency")
plt.show()
```

```python
numeric_columns=iris.select_dtypes(include=['float64','int64'])
corr_matrix=numeric_columns.corr()
print(corr_matrix)
plt.figure(figsize=(5,3))
sns.heatmap(corr_matrix,annot=True,cmap="Spectral",fmt=".2f")
plt.title("CORRELATIN MAP OF Price  DATASET")
plt.show()
```
